```python
%matplotlib inline 

import pandas as pd 
import seaborn as sns 

sns.set_context('poster')
sns.set_style('whitegrid')
```

```python
data = pd.read_csv('local.csv')
```

```python
data.dtypes
```

```python
data.describe()
```

```python
g = sns.barplot('host', 'throughput (msg/sec)', hue="trace", data=data)
g.set_xlabel('host (vm vs. metal)')
g.set_ylabel('mean client throughput (writes/sec)')
g.set_title('Single Local Client Throughput')
```
